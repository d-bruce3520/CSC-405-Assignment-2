We are going to study the survival rate of passengers on titanic and what variables affected survival.

***1. Load the dataset in `titanic.xls`. It contains data on all the passengers that travelled on the Titanic. Show the head of 10 rows (10 pts)***

```python
import pandas as pd
import numpy as np

titanic_file = pd.ExcelFile('titanic.xls')
titanic_file
```

```python
titanic_data = titanic_file.parse()

titanic_data.head(10)
```

*** 2. Use the `groupby` method to calculate the proportion of passengers that survived by sex. (20 pts)*** 

```python
titanic_group1 = titanic_data.groupby('sex')
#Since survival is shown based on binary 0/1 value, the total survivors can be determined by summing the survived column
survivors_by_sex = titanic_group1['survived'].sum()
print(survivors_by_sex)
```

```python
#Find and display total passengers by group to help with conclusions
passengers_by_sex = titanic_group1['sex'].count()
print(passengers_by_sex)
```

Based on grouping the data by sex, more twice as many female passengers survived than male passengers.

*** 3. Calculate the same proportion, but by pclass and sex. (30 pts)*** 

```python
titanic_group2 = titanic_data.groupby(['pclass', 'sex'])
survivors_by_pclass = titanic_group2['survived'].sum()
print(survivors_by_pclass)
```

```python
#Find and display total passengers by group to help with conclusions
passengers_by_pclass = titanic_group2['pclass'].count()
print(passengers_by_pclass)
```

Grouping by sex and pclass shows that most survivors were of pclass 1, then pclass 3, and finally pclass 2. Significantly more female passengers survived than male passengers in pclass 1 and 2, while in pclass 3 only 31 more female passengers survived than males, a much lesser difference than the other two pclasses. Based on total passengers of each pclass, the pvlasses were prioritized first class 1, the  class 2, then class 3. As despite pclass 3 having the 2nd most survivors, less than half of the total pclass 3 passengers survived while a much greater portion of the pclass 1 and 2 passengers survived.

*** 4. Create age categories: children (under 14 years), adolescents (14-20), adult (21-64), and senior(65+), and calculate survival proportions by age category, pclass and sex. (30 pts)***

```python
#Create function to return age group based on age value of row in dataframe.
def assign_age_group(row):
    if row['age'] < 14:
        return 'children'
    if 14 <= row['age'] <= 20:
        return 'adolescents'
    if 21 <= row['age'] <= 64:
        return 'adult'
    if row['age'] > 64:
        return 'senior'
    return 'N/A'
```

```python
#Create age group column and fill with values
titanic_data['age group'] = titanic_data.apply(assign_age_group, axis=1)
#test that new column was created and filled
titanic_data.head(10)
```

```python
titanic_group3 = titanic_data.groupby(['age group', 'pclass', 'sex'])
survivors_by_age_group = titanic_group3['survived'].sum()
print(survivors_by_age_group)
```

Grouping by age group, pclass and sex shows that significantly more adults survived than any other age group regardless odf sex and pclass. Only 2 seniors, one man and one woman both of pclass 1, survived. With the exception of adults of pclass 3 and children of pclass 1, the number of female survivors was greater than or equal to the number of male survivors. The largest group of survivors is adult women of pclass 1.

```python
#Find and display total passengers by group to help with conclusions
passengers_by_age_group = titanic_group3['age group'].count()
print(passengers_by_age_group)
```

The results of these three groupings show that female passengers were largely prioritized over male passengers. Additionally it seems that younger passengers were also prioritized, the majority of survivors being adults is due to the adult age group making up the majority of passengers. Lastly it seems that passengers were prioritized by pclass, with passengers of pclass 1 having the most survivors, both total and relative to the number of passengers of that group. Based on the total passengers significantly more passengers of pclasses 1 and 2 survived than pclass 3.
